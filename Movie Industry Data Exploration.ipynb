There has been several movies over the years that have become the most popular movies of all time. Some movies were produced by the top studios such as Marvel and Warner Brothers. The gross earnings of the top movies have earned up to millions and billions of revenue. This data exploration will focus on the factors of a movie that influences gross earnings. These factors can include actors, production studios, etc. 

Research Question: What factors have a major influence towards the gross earnings of popular movies?

Hypothesis: High gross earning movies have a high budget amount compare to movies that have a low budget. 

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read in CSV file
df = pd.read_csv('movies.csv')
df.head()
```

# Data Cleaning 
The data cleaning process composes:
 
 - viewing the percents of empty cells per column.
 

- dropping rows containing empty cells
 
 
 
 - converting several column's data types
 
 
 - extracting values from existing columns via regex

```python
# Find Missing Data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
df.dtypes
```

```python
# Change data type for budget and gross columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Correcting the year column by using regex to extract the year due to 'year' and 'released' columns being similar
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
# Sorting the dataframe by top grossing films
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates in the dataframe

df.drop_duplicates()
```

# Data Visualization 

Data visualization composes of:

- Scatterplots of budget vs gross earnings


- Heat map of correlation matrix of numerical and categorical values 

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title("Budget vs Gross Earnings")
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earning')

plt.show()

```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
df.corr()
```

```python
# High correlation between budget and gross

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

Both the heat map and scatterplot shows how a high budget investment for a movie results in a high gross earnings. The correlation value of budget vs earnings is 0.74. With a high budget investment for a movie will produce a high quality movie for viewers to enjoy. 

```python
# Assigning code values to categorical columns to create a correlation matrix
df_numerical = df.copy()

for col_name in df_numerical.columns:
    if(df_numerical[col_name].dtype == 'object'):
        df_numerical[col_name] = df_numerical[col_name].astype('category')
        df_numerical[col_name] = df_numerical[col_name].cat.codes
        
df_numerical.head()  
```

```python
correlation_matrix = df_numerical.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for all Variables")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerical.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

High amounts of votes for a movie have a influence to gross earnings. The viewers have a major impact to the gross earnings because the enjoyment of the movie allows viewers to vote on the movie.
